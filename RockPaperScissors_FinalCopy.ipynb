```python
def run_game():
    # Generate Computer throw - (Rock paper scissors) 
    from random import shuffle

    def my_shuffle(ls):
        shuffle(ls)
        return ls

    c = ['rock','paper','scissors']
    shuffle(c)

    # test throw
    compThrow = c[0]
    #c[0]
    #------------------------------------------------#


    # convert user numeric input to string for my preference

    def get_userThrow():
        userThrow, x = 0, 0
        while userThrow == 0:   
            x = input("""Chose your throw (enter 1, 2 or 3):
            1 - Rock
            2 - Paper               
            3 - Scissors
                ----------->  """)
            if x == '1':
                userThrow = 'rock'
            elif x == '2':
                userThrow = 'paper'
            elif x == '3':
                userThrow = 'scissors'
            else:
                print(f'Sorry, {x} is not a valid selection.\n\n')
        return userThrow



    userThrow = get_userThrow()
    print(f'You chose {userThrow}. Here we go...')

    #------------------------------------------------#

    # Check user throw against computer throw, output user results

    def throw_compare(comp = 'none', user = 'none'):
        # comp TIE situation--------------------------------------
        if comp == user:
            outcome = 'TIE'
        # comp LOSE situation--------------------------------
        elif (comp == 'rock' and user == 'paper') or (comp == 'paper' and user == 'scissors') or (comp == "scissors" and user == 'rock'):
            outcome = 'YOU WIN'
        # comp WIN situation ( not needed with else but in here for error testing )
        elif (comp == 'rock' and user == 'scissors') or (comp == 'paper' and user == 'rock') or (comp == 'scissors' and user == 'paper'):
            outcome = 'YOU LOSE'
   
        return outcome
 
 
    outcome = throw_compare(compThrow,userThrow)
    print('\n\n{0:<10} |{1:^10}| {2:>10}' .format(userThrow, 'VS', compThrow))
    print(' {0:.>32} \n {1:^32} \n {0:.>32}' .format('',outcome,''))
    
    return outcome
```

```python
def beginning(hold):
    
    win_history = [hold]
    correct_responses = ['y','Y','n','N']
    x = 0
    start = 0
    
    while start not in correct_responses:
        start = input('Hello, would you like a game of Rock, Paper, Scissors? (Y/N):  ')
        if start == 'y' or start == 'Y':
            x = 'y'
            while x == 'y':
                if x == 'Y' or x == 'y':
                    
                    y = run_game() #this should make y = outcome returned by reun_game
                    win_history.append(y) #this should add the outcome to the win history returned to the begin call
            
                    again = 0
                    while again not in correct_responses:
                        again = input("\nWould you like to play again? (Y/N):  ")
                        if again == 'Y' or again == 'y':
                            x = 'y'
                            continue
                        elif again == 'n' or x == 'N':
                            print("\n**Thank you for playing!**\n Here are the results of your game: \n\n\t")
                         
                            games = len(win_history)
                            i = 0
                            game_number = 0
                            for i in range(games):
                                print("Game", game_number, ": ", win_history[i])
                                i += 1
                                game_number += 1
                            
                            wins = win_history.count('YOU WIN')
                            loses = win_history.count('YOU LOSE')
                                
                            if wins > loses:
                                print(f"\nYou won this round of games\n\t {wins} to {loses} out of {games-1} games!\nCONGRATS!!")
                            elif wins < loses:
                                print(f"\nYou lost this round of games\n\t {wins} to {loses} out of {games-1} games!\n.....Better luck next time ;)")
                            else:
                                print(f"Overall, it was a tie with a total of {games-1} games.")
                       
                            x = 0
                        else:
                            print("...........Please enter only 'Y' for yes or 'N' for no\n")
                elif x == 'N' or x == 'n':
                    print("Thank you, maybe next time we can play. \n\n\n")
                else:
                    print("...........Please enter only 'Y' for yes or 'N' for no\n")
        else:
            print('Thank you, maybe next time!')
        
    
```
