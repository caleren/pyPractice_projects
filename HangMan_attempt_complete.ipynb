```python
def print_hangman(correct_guesses,error_man):
    print_man = error_man
    print(' ___')
    print('|   |')
    print(f'|   {print_man[0]}')
    print(f'|  {print_man[2]}{print_man[1]}{print_man[3]}')
    print(f'|  {print_man[4]} {print_man[5]}')
    
    for x in range(len(correct_guesses)):
        print(correct_guesses[x], end = ' ')


def game_query():
    x = 'b'
    while not (x == 'y' or x == 'n'):
        x = input("Would you like to play hangman? ('Y/N')")
        x = x.lower()
        if (x == 'y'):
            run_game()
        elif (x == 'n'):
            print('Have a nice day. Goodbye.')
            break
        else:    
            print("Please enter only 'Y' or 'N'")
        
        
def run_game():
    from IPython.display import clear_output


    word = 'randomword'
    correct_letters = [char for char in word]
    correct_guesses = ['_' for char in word]
    alpha_list = alphabet = ['a','b','c','d','e','f','g','h','i','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
    used_letters = []
    error_count = 0
    hang_man = ['O','|','/','\\','/','\\']
    error_man = [' ',' ',' ',' ',' ',' ']
    
    while error_count < 6:
        print_hangman(correct_guesses, error_man)
        guess = input("Guess a letter: ")
        guess = guess.lower()                                     #make uppercase letters lowercase
        if (guess in alpha_list):                                 #check if guess is a valid letter in alphabet
            if (not guess in used_letters):                       #check if guess has been used
                used_letters.append(guess)                        #add guess to used letters so won't be used again
                if guess in correct_letters:
                    clear_output()
                    print(f'{guess} is correct!')
                    for index in range(len(correct_letters)):     #find index of matching letters 
                        for char in correct_letters:
                            if guess == correct_letters[index]:
                                correct_guesses[index] = guess
                                if correct_guesses == correct_letters:
                                    clear_output()
                                    print('YOU WIN!')
                                    print(f'The word was {word}\n\n')
                                    game_query()
            
                    
            
                else:
                    clear_output()
                    print(f'{guess} is NOT in this word')
                    error_man[error_count] = hang_man[error_count]
                    error_count +=1
                
        
            else:
                clear_output()
                print(f'You already tried {guess}. Guess again.\n\n')
        else:
            clear_output()
            print(f'{guess} is not a letter. Try again.\n\n')
            
    clear_output()
    print_hangman(correct_guesses, error_man)
    print(f'\n\n Sorry, you lost. Correct word was {word}')        
    game_query()


game_query()
```
